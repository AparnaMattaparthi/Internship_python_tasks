```python
# Import required modules
import os
import shutil

# -------------------------------
# STEP 1: Writing to a text file
# -------------------------------
try:
    with open("sample.txt", "w") as file:
        file.write("Hello, this is a sample file.\n")
        file.write("This file is created using Python.\n")
    print("File written successfully.")

except Exception as e:
    print("Error while writing file:", e)
```

```python
# Import required modules
import os
import shutil

# -------------------------------
# STEP 2: Reading from a file
# -------------------------------
try:
    with open("sample.txt", "r") as file:
        content = file.read()
        print("\nFile Content:\n", content)

except Exception as e:
    print("Error while reading file:", e)
```

```python
# Import required modules
import os
import shutil

# -------------------------------
# STEP 3: Rename the file
# -------------------------------
try:
    os.rename("sample.txt", "renamed_sample.txt")
    print("File renamed successfully.")

except Exception as e:
    print("Error while renaming file:", e)
    
    
```

```python
# Import required modules
import os
import shutil

# -------------------------------
# STEP 4: Move file to a folder
# -------------------------------
try:
    # Create folder if not exists
    if not os.path.exists("test_folder"):
        os.mkdir("test_folder")

    shutil.move("renamed_sample.txt", "test_folder/renamed_sample.txt")
    print("File moved successfully.")

except Exception as e:
    print("Error while moving file:", e)
```

```python
# Import required modules
import os
import shutil

# -------------------------------
# STEP 5: Delete file
# -------------------------------
try:
    os.remove("test_folder/renamed_sample.txt")
    print("File deleted successfully.")

except Exception as e:
    print("Error while deleting file:", e)
```
